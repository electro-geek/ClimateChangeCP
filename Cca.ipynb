```python
loc_dict = dict()
count = 0
for i in location_column:
    loc_dict[i] = count
    count += 1
print(loc_dict)
```

```python
for i in dataset['LOCATION']:
    temp = loc_dict[i]
    dataset = dataset.replace(i,temp)
```

```python
#so we can remove columns which are constant
dataset = dataset.drop(['INDICATOR', 'FREQUENCY', 'Flag Codes', 'index'], axis = 1)
```

```python
dataset = dataset.drop(['index'], axis = 1)
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset, columns = ['SUBJECT'], drop_first=True)
```

```python
dataset.head(20)
```

```python
dataset = pd.get_dummies(dataset, columns = ['MEASURE'], drop_first=True)
```

```python
dataset.head(20)
```

```python
dataset.isnull().sum()
```

```python
#taking dependent and independent variables
X = dataset.drop(['Value'], axis=1)
```

```python
X.head()
```

```python
y = dataset['Value']
```

```python
y.head()
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl', 'wb'))
```

```python
lin_reg.fit(X_train,y_train)
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(mean_squared_error(y_test, y_pred))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import Ridge
rdg_reg = Ridge(max_iter = 10000, )
```

```python
params = {
    'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20],
}
```

```python
from sklearn.model_selection import GridSearchCV
rg_model = GridSearchCV(rdg_reg, param_grid = params, scoring = 'accuracy', cv = 10)
```

```python
rg_model.fit(X_train,y_train)
```

```python
y_pred = rg_model.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_pred,y_test))
```

```python
import pickle
pickle.dump(rg_model, open('regmodel.pkl', 'wb'))
pickle_model = pickle.load(open('regmodel.pkl', 'rb'))
```
